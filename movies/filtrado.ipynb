```python
import pandas as pd
```

```python
df_movies = pd.read_csv('IMDb movies.csv', low_memory=False)
df_movies.columns
```

```python
df_rating = pd.read_csv('IMDb ratings.csv',low_memory=False)
df_rating.columns
```

```python
df_movies = df_movies[['imdb_title_id', 'title', 'year', 'genre', 'country']]
```

```python
df_rating = df_rating[['imdb_title_id', 'total_votes', 'mean_vote']]
```

```python
#concatenacion 
df1 = pd.DataFrame ({'id': ['A', 'B', 'C', 'D'],
        'age': [30, 23, 25, 22]})
df2 = pd.DataFrame({'id': ['E', 'F', 'G', 'H'],
        'age': [40, 21, 19, 24]})
```

```python
pd.concat([df1, df2], axis=0, ignore_index=True, sort=True)
```

```python
#extraer 50% de la muestra original del dataframe (df_movies)
df_sample = df_movies.sample(frac=0.5)
```

```python
df_sample.shape
```

```python
#concatenar df_movies y df_rating
df_concat = pd.concat([df_movies, df_rating], axis=1, ignore_index=True, sort=True)
df_concat
```

```python
#inner join dos formas de hacer lo mismo
df_movies.merge(df_rating, on='imdb_title_id')
pd.merge(df_movies, df_rating, on='imdb_title_id')
```

```python
#outer_join (full join) 
df1.merge(df2, on='id', how='outer')
```

```python
#full join de df_movies y df_ratings
df_movies.merge(df_rating, on='imdb_title_id', how='outer')
```

```python
df_movies.merge(df_rating, on='imdb_title_id', how='outer', indicator=True).query("_merge=='left_only' or _merge=='right only'")
```

```python
#left join
df_movies.merge(df_rating, on='imdb_title_id', how='right').shape
```
