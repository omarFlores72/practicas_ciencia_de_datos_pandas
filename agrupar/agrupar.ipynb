```python
import pandas as pd
```

```python
df_cars = pd.read_csv('Car_sales.csv')
```

```python
df_cars_frame = df_cars[['Manufacturer', 'Sales_in_thousands', 'Vehicle_type', 'Price_in_thousands', 'Engine_size', 'Horsepower', 'Fuel_capacity']]
```

```python
#cantidad de elementos unicos
df_cars_frame.nunique()
```

```python
#agregar fucniones con mtodo agg
#df_cars_frame.select_dtypes(include='number').agg('sum')
df_cars.select_dtypes(include='number').agg('mean')
```

```python
df_cars_frame.select_dtypes(include='number').agg(['sum', 'mean'])
```

```python
#aplicar 1 o mas funciones a columnas seleccionadas, en este caso, mediana y suma
df_cars_frame.agg({'Sales_in_thousands':['sum', 'mean']
                    ,'Price_in_thousands':['sum', 'max']})
```

```python
df_cars[['Sales_in_thousands', 'Price_in_thousands']].agg('sum', axis=0) #1 es columna
```

```python
#cambiar nombres a filas
df_cars.agg(x=('Sales_in_thousands', 'sum'),
            y=('Price_in_thousands', 'sum'))
```

ESTRATEGIA Split-Appy-Combine agrupar en categorias

```python
#split separar la data
#df_cars['Vehicle_type'].value_counts()

car_filter = df_cars['Vehicle_type']=='Car'
passenger_filter = df_cars['Vehicle_type']=='Passenger'
```

```python
#apply
car_average = df_cars[car_filter]['Sales_in_thousands'].mean()
passenger_average = df_cars[passenger_filter]['Price_in_thousands'].mean()
```

```python
#combine
pd.DataFrame({'Vehicle_type':['Car', 'Passenger'],
            'Sales_in_thousands':[car_average, passenger_average]})
```

METODO GROUPBY Y METODO AGG JUNTOS

```python
df_cars.groupby('Vehicle_type').agg(['min', 'max'])
```

```python
df_cars.groupby('Vehicle_type').agg( min_engine_size = ('Engine_size', 'min'),
                                    max_horsepower = ('Horsepower', 'max'))
```

```python
#calcular la suma de las ventas y promedio de precios
df_cars.groupby('Manufacturer').agg(sum_sales = ('Sales_in_thousands', 'sum'),
                                    average_price =('Price_in_thousands', 'mean'))
```

```python
#lambda
df_cars.groupby('Manufacturer').agg('sum',numeric_only=True).apply(lambda x:x*1000)[['Sales_in_thousands', 'Price_in_thousands']]
```

```python
#restar valores de la media a cada grupo
df_cars.groupby('Manufacturer').agg('sum',numeric_only=True).apply(lambda x:x-x.mean())['Engine_size']
```
