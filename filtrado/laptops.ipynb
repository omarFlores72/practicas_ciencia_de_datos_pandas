CREAR COLUMNA CONDICIONAL CON 1 O MAS OPCIONES

```python
import numpy as np
```

```python
#crear array basado en multiples niveles de precio 
#crear condiciones y valores
condiciones = [
    df_laptops['Price_euros']>3000,
    (df_laptops['Price_euros']>2000) & (df_laptops['Price_euros']<=3000),
    (df_laptops['Price_euros']>800) & (df_laptops['Price_euros']<=2000),
    df_laptops['Price_euros']<=800
]
valores = ['Muy caro', 'Caro', 'barato', 'Muy Barato']
```

```python
df_laptops['Niveles_precio'] = np.select(condiciones, valores, default='Sin categorÃ­a')
```

```python
df_laptops
```

```python
df_laptops.value_counts('Niveles_precio')
```

```python
#array de tamaÃ±o pantalla muy grande>16, grande>14, pequeÃ±o>12, muy pequeÃ±o<12 
condiciones = [
    df_laptops['Inches']>16,
    (df_laptops['Inches']>14) & (df_laptops['Inches']<=16),
    (df_laptops['Inches']>14) & (df_laptops['Inches']<=12),
    df_laptops['Inches']<=12
]
valores = ['Muy grande', 'Grande', 'pequeÃ±o', 'Muy pequeÃ±o']
```

```python
df_laptops['Nivel_tamaÃ±o'] = np.select(condiciones, valores, default='Sin categorÃ­a')
```

```python
df_laptops.value_counts('Nivel_tamaÃ±o')
```

```python
#por defecto keep='first
duplicated_fisrt = df_laptops.duplicated('Company')
```

```python
#eliminar duplicados en 2 o mas columnas
df_laptops.drop_duplicates(['Company'])[['Company', 'Price_euros']]
```

```python
#mas barato. inplace guarda los daos en lugar de arrojar una copia
df_laptops.drop_duplicates(['Company'], keep='last',inplace=True, ignore_index=True)
df_laptops[['Company', 'Price_euros']]
```
