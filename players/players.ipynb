SAMPLE AYUDA A EXTRAER UNA MUESTRA ALEATORIA DE UN DATAFRAME

```python
#EXTRAER 10 ELEMENTOS ALEATORIOS
df_player['nationality'].sample(50, random_state=100)
```

```python
#extraer 20% de la muestra aleatoria de un dataframe
df_player.sample(frac=0.35)
```

```python
#incrementar la muestra -> frac>1 (obs: el paramero replace tiene que ser True cuando el frac>1)
df_player.sample(frac=2, replace=True, random_state=100) #random_state hace que ya no cambie aleatoriamente la muestra
```

```python
#metodo query filtrar jugadores maoyores a 34 años
df_player.query("age>=18 and nationality=='Mexico'")  
```

```python
#convertir el ditpo de dato dob que es object a datetime
df_player['dob'] = df_player['dob'].astype('datetime64[ns]')
df_player['dob'].dt.year
```

```python
df_player.query("dob.dt.year == 1998")
```

```python
import numpy as np
df_player['age'].apply(np.sqrt)
```

```python
#funcion para indice de masa corporal
def calcular_imc(row):
    return row['weight_kg']/((row['height_cm']/100)**2)
```

```python
df_player.apply(calcular_imc, axis=1)
```

```python
#funcion basica
def sum_values(a,b):
    x = a+b
    return x
print(sum_values(8,52))
```

```python
#funcion lambda
sum_values_lambda = lambda a,b:a+b
```

```python
sum_values_lambda(2,5)
```

```python
#funcion lambda para convertir cm a metros
df_player['height_cm'].apply(lambda x:x/100)
#alternativa
df_player['height_cm']/100
```

```python
#funcionlambda para convertir serie "long name" a mayuscula
df_player['long_name'].apply(lambda x:x.upper())
#alternatia
df_player['long_name'].str.upper()
```

```python
#obtener añode la serie dob
df_player['dob'].apply(lambda x:x.year)
```
