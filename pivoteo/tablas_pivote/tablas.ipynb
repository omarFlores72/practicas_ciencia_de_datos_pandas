```python
import pandas as pd
```

```python
df_population = pd.read_csv('population_total.csv')
df_population 
```

```python
#hacer tabla pivote
df_pivot = df_population.pivot(index='year', columns='country', values='population')
```

```python
df_pivot = df_pivot[['Mexico', 'India', 'Argentina', 'Spain', 'United States', 'China', 'Peru']]
```

grafico de linea con LinePlot()

```python
df_pivot.plot(kind='line', xlabel='año', ylabel='población', title='polacion 1955-2020', figsize=(8,4))
```

grafico de caja BARPLOT (grafico de caja)

```python
#seleccionar solo un año
df_pivot_2020 = df_pivot[df_pivot.index.isin([2020])]
```

```python
#.T para invertir filas y columnas
df_pivot_2020 = df_pivot_2020.T 
```

```python
#hacer barplot
df_pivot_2020.plot(kind='bar')
```

```python
#seleccionar unos años
df_pivot_sample = df_pivot[df_pivot.index.isin([1980, 1990, 2000, 2020])]
df_pivot_sample.plot(kind='bar')
```

```python
#cambando nombre a la columna
df_pivot_2020 = df_pivot_2020.copy()
df_pivot_2020.rename(columns={2020.0: '2020'}, inplace=True)
```
