```python
import pandas as pd
import statsmodels.api as sm
```

```python
df_boston = pd.read_csv('/home/omar/pandas/csv/BostonHousePrices.csv')
```

```python
df_boston.describe()
```

```python
#regresion lineal: algoritmo para modelar relacion ente 2 o mas variables
y = df_boston['Value']
x = df_boston['Rooms']
```

```python
df_boston.plot(kind='scatter',
               x='Rooms',
               y='Value')
```

```python
x = sm.add_constant(x)
lm = sm.OLS(y, x).fit() #modelo ajustado
```

```python
lm.predict(x)
```

```python
lm.summary()
```

```python
#ecuacion de regresion lineal
#coef rooms
#coesf constants
#ecuacuon lineal
y_pred = 9.1021*x['Rooms'] -34.6706
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#graficar puntos
plt.figure(figsize=(6, 4), tight_layout=True)
sns.lineplot(x=x['Rooms'],y=y)
#graficar la linea
sns.lineplot(x=x['Rooms'],y=y_pred, color='black')
#axis
plt.xlim(0)
plt.ylim(0)
plt.savefig('linear_regression')
plt.show()
```

```python
#regresion lineal multiple
from sklearn import linear_model
```

```python
y = df_boston['Value']
x = df_boston[['Rooms', 'Distance']]
```

```python
#ajustar el modelo
lm = linear_model.LinearRegression()
lm.fit(x, y)
```

```python

lm.predict(x)[:10]
```

```python
#r2 score
lm.score(x, y)
```

```python
#coeficiente
lm.coef_
```
