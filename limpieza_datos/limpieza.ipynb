```python
import pandas as pd
```

```python
df_titles = pd.read_csv('netflix_titles.csv')
```

```python
df_titles.shape
```

MANEJO DE DATA FALTANTE

```python
df_titles.isnull().sum().sort_values(ascending=False)
```

```python
# %filas faltantes
for column in df_titles.columns:
    porcentaje = df_titles[column].isnull().mean()
    print(column + ':' + str(round(porcentaje*100, 2)))
```

```python
df_titles.drop('director', axis=1)
```

```python
df_no_director = df_titles[df_titles['director'].isnull()].index
df_titles.drop(df_no_director, axis=0)
```

```python
#reemplazarpor la media , mediana y moda
moda = ''.join(df_titles['rating'].mode())
df_titles['rating'].fillna(moda, inplace=True)
```

```python
df_titles.loc[5000]
```

```python
#reemplazar numeerico arbitrario con .fillna()
df_titles['duration'].fillna('0',inplace=True)
```

```python
df_titles.isnull().sum()
```

```python
#backward y forward
df_titles.fillna(method='bfill')
df_titles.fillna(method='ffill')
```

```python
#extraer data de una columna
df_movie = df_titles[df_titles['type']=='Movie']
df_movie['minute'] = df_movie['duration'].str.split(expand=True)[0].astype(int) #split por defecto serpara por espacio en blanco
```

```python
# split se usa para dividir cadenas dentro de una Serie o columna de un DataFrame, basándose en un delimitador especificado
df_movie['date_added'].str.split(pat=',', expand=True) #pat = patron
```

identificar valores atipicos
usar histograma para identificar valores atipicos dentro de una data numerica

```python
#histograma
df_movie['minute'].plot(kind='hist', bins=20)  #bins= numero de barras
```

```python
df_movie['minute'].value_counts(bins=10).sort_index()
```

```python
df_movie[~(df_movie['minute']>31.2) & (df_movie['minute']<218)].value_counts().sort_values().sum() #~ para negar
```

```python
#boxplot para valores atipicos 
df_movie['minute'].plot(kind='box', vert=False, color='blue', figsize=(15,7))
```

```python
#revision de valores estadisticos
#IQR = Q3 -Q1
min_boxplot = 87 - 1.5*(114-87)
#Q1 - 1.5*IQR
#Q3 + 1.5*IQR
max_boxplot = 114 + 1.5*(114-87)
print(min_boxplot)
print(max_boxplot)
#df_movie['minute'].describe()
```

```python
#manejo valores atipicos
df_movie[~(df_movie['minute']>46.5) & (df_movie['minute']<154.5)].value_counts().sort_values().sum()
```
